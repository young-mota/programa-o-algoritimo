```python
# estrutura de decisões

# escrever um programa que recebe o nome e a idade do usuario e verifica se o usuario é criança, adolecente ou adulto

#idade <= 17 anos é criança
# <= 59 anos é adolecente
#idoso

#entrada
nome = str(input("digite o seu nome"))
idade = int(input("digite sua idade"))
#processamento/ saida

#saida
if idade <= 17:
  print(f"{nome}, você tem {idade} anos de idade, você é criança")
elif idade <= 59:
  print(f"{nome}, você tem {idade} anos de idade, você é adulto")
else:
  print(f"{nome}, você tem {idade} anos de idade, você é idoso")
```

```python
valor_compra = float(input("Digite o valor da compra: R$"))
#entrada
if valor_compra <= 100:
    valor_final = valor_compra
    desconto = 0
elif 101 <= valor_compra <= 500:
    desconto = 0.10
    valor_final = valor_compra * (1 - desconto)
else:
    desconto = 0.20
    valor_final = valor_compra * (1 - desconto)
#saida
print(f"O valor da compra foi: R${valor_compra:.2f}")
if desconto > 0:
    print(f"Desconto aplicado: {desconto*100:.0f}%")
print(f"O valor final a ser pago é: R${valor_final:.2f}")
```

```python
print("Verificação de Triângulo")

lado1 = float(input("Digite o comprimento do primeiro lado: "))
lado2 = float(input("Digite o comprimento do segundo lado: "))
lado3 = float(input("Digite o comprimento do terceiro lado: "))

# Verifica a condição de existência de um triângulo
if (lado1 + lado2 > lado3) and \
   (lado1 + lado3 > lado2) and \
   (lado2 + lado3 > lado1):
    print("Os valores podem formar um triângulo válido.")
else:
    print("Os valores NÃO podem formar um triângulo válido.")
```

```python
try:
    energia = int(input("Digite a energia do personagem (de 0 a 100): "))

    #precossamento/saida
    if not (0 <= energia <= 100):
        print("Erro: A energia deve estar entre 0 e 100.")
    elif energia == 0:
        print("Classificação: Morto")
    elif 1 <= energia <= 30:
        print("Classificação: Crítico")
    elif 31 <= energia <= 70:
        print("Classificação: Normal")
    elif 71 <= energia <= 100:
        print("Classificação: Saudável")
except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
```

```python
try:
    media_final = float(input("Digite a média final do aluno: "))

    if 0 <= media_final <= 10: # Assuming a valid grade is between 0 and 10
        if media_final >= 7.0:
            print("Classificação: Aprovado")
        elif 5.0 <= media_final <= 6.9:
            print("Classificação: Recuperação")
        else:
            print("Classificação: Reprovado (abaixo de 5.0)") # Added for completeness
    else:
        print("Erro: A média deve ser um valor entre 0 e 10.")
except ValueError:
    print("Entrada inválida. Por favor, digite um número para a média.")
```

```python
try:
    salario = float(input("Digite o salário do funcionário: R$"))
    pontuacao = int(input("Digite a pontuação de desempenho (0 a 10): "))

    bonus = 0.0

    if not (0 <= pontuacao <= 10):
        print("Erro: A pontuação de desempenho deve ser entre 0 e 10.")
    else:
        if pontuacao < 5:
            bonus_percent = 0.0
        elif 5 <= pontuacao <= 7:
            bonus_percent = 0.05
        elif 8 <= pontuacao <= 9:
            bonus_percent = 0.10
        else: # pontuacao == 10
            bonus_percent = 0.15

        bonus = salario * bonus_percent
        salario_final = salario + bonus

        print(f"Salário Base: R${salario:.2f}")
        print(f"Pontuação de Desempenho: {pontuacao}")
        print(f"Valor do Bônus: R${bonus:.2f}")
        print(f"Salário Final com Bônus: R${salario_final:.2f}")

except ValueError:
    print("Entrada inválida. Por favor, digite números válidos para salário e pontuação.")
```
