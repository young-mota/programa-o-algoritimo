escrever um programa que calcula a média das notas do aluno

entrada: nome, idade, e as 4 notas do bimestre

processamento: calcular a media

saida: mostrar o nome do aluno e a média óbtid.

```python
#entrada
nome = str = (input("digite seu nome: "))
idade = float(input("digite sua idade: "))
nota = float(input("digite a primeira nota:"))
nota2 = float(input("digite sua segunda nota:"))
nota3 = float(input("digite sua 3 nota:"))
nota4 = float(input("digite sua 4 nota:"))

#processamento

media = (nota+ nota2+ nota3+ nota4) / 4

#saida

print(f"olá {nome}")
print(f"você tem {idade} anos")
print(f"sua média é de {media}")
```
