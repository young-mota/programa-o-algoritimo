```python
#entrada

nome = input("Digite o nome do aluno: ")
nota = int(input("Digite a nota do aluno: "))

#processamento e saída

if nota >= 7:
    print (f"O aluno {nome} aprovado com a nota {nota}")
elif nota >= 5: # Corrigido: indentação e uso de 'elif'
    print(f"O aluno {nome} está em recuperação pela nota ser {nota}")
else:
    print(f"O aluno {nome} foi reprovado por estar com a nota {nota}")
```

```python
 estrutura de decisões

# escrever um programa que recebe o nome e a idade do usuario e verifica se o usuario é criança, adolecente ou adulto

#idade <= 17 anos é criança
# <= 59 anos é adolecente
#idoso

#entrada
nome = str(input("digite o seu nome"))
idade = int(input("digite sua idade"))
#processamento/ saida

#saida
if idade <= 17:
  print(f"{nome}, você tem {idade} anos de idade, você é criança")
elif idade <= 59:
  print(f"{nome}, você tem {idade} anos de idade, você é adulto")
else:
  print(f"{nome}, você tem {idade} anos de idade, você é idoso")
```

```python
#entrada

numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))
operacao = input("Digite a operação (+, -, *, /): ")

if operacao == "+":
    resultado = numero1 + numero2
    print(f"O resultado da soma é: {resultado}")
elif operacao == "-":
    resultado = numero1 - numero2
    print(f"O resultado da subt é: {resultado}")
elif operacao == "*":
    resultado = numero1 * numero2
    print(f"O resultado da mult é: {resultado}")
elif operacao == "/":
    resultado = numero1 / numero2
    print(f"O resultado da divisão é: {resultado}")
```

```python
#entrada

dia = int(input("Digite um número de 1 a 7: "))

#processamento e saída

if dia == 1:
    print("Domingo")
elif dia == 2:
    print("Segunda-feira")
elif dia == 3:
    print("Terça-feira")
elif dia == 4:
    print("Quarta-feira")
elif dia == 5:
    print("Quinta-feira")
elif dia == 6:
    print("Sexta-feira")
elif dia == 7:
    print("Sábado")
else:
    print("Esse é o dia maluco vai se tratar")
```
