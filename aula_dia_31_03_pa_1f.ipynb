```python
# estrutura de decisões

# escrever um programa que recebe o nome e a idade do usuario e verifica se o usuario é criança, adolecente ou adulto

#idade <= 12 anos é criança
# <= 18 anos é adolecente
#adulto

#entrada
nome = str(input("digite o seu nome"))
idade = int(input("digite sua idade"))
#processamento/ saida


if idade <= 12:
  print(f"{nome}, você tem {idade} anos de idade, você é criança")
elif idade <= 18:
  print(f"{nome}, você tem {idade} anos de idade, você é adolecente")
else:
  print(f"{nome}, você tem {idade} anos de idade, você é adulto")
```

```python
#em um jogo um personagem possui uma variavel chamada energia o programa deve: se a energia for maior que 70 - forte, se a energia estiver entre 30 e 70 - cansado, se a energia estiver menor que 30- fraco.

#entrada
energia= int(input("digite a energia do seu personagem"))

#processamneto/saida

if energia <30:
  print(f"seu personagem está fraco")
elif energia >=30 and energia <=70:
  print(f"seu personagem está cansado")
else:
  print(f"seu personagem está forte")
```

```python
#escrever um programa que classsfica os tipos de triangulo; e valida se é um triangulo
#validar se os lados digitados pode formar um triangulo
#lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1

#tipos de triangulo
#Equilatero - todos os lados iguais - ==
#Escaleno - todos os lados diferentes - !=
#isóceles - dois lados iguais

#entrada
lado1 = int(input("Digite o comprimento do primeiro lado: "))
lado2 = int(input("Digite o comprimento do segundo lado: "))
lado3 = int(input("Digite o comprimento do terceiro lado: "))

#processamento/saida

# Validação se pode formar um triângulo
if (lado1 + lado2 > lado3) and (lado1 + lado3 > lado2) and (lado2 + lado3 > lado1):
  print("Os lados digitados podem formar um triângulo.")
  # Classificação do triângulo
  if lado1 == lado2 and lado2 == lado3:
    print("Este é um triângulo Equilátero (todos os lados iguais).")
  elif lado1 != lado2 and lado1 != lado3 and lado2 != lado3:
    print("Este é um triângulo Escaleno (todos os lados diferentes).")
  else:
    print("Este é um triângulo Isósceles (dois lados iguais).")
else:
  print("Os lados digitados NÃO podem formar um triângulo.")
```
